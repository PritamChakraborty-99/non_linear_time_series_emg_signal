```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg

# Generate or load your time series data
# For demonstration purposes, let's generate a simple sinusoidal signal

signal = emg_bcps

# Function to fit an autoregressive model and calculate AIC
def fit_ar_model(data, lag):
    model = AutoReg(data, lags=lag)
    model_fit = model.fit()
    return model_fit.aic

# Range of lags to consider
max_lag = 100
lags = np.arange(1, max_lag)

# Fit autoregressive models for different lags and calculate AIC
aic_values = [fit_ar_model(signal, lag) for lag in lags]

# Plot AIC values
plt.plot(lags, aic_values, marker='o')
plt.title('AIC vs. Lag for Autoregressive Model (Biceps)',fontweight='bold')
plt.xlabel('Lag',fontweight='bold')
plt.ylabel('AIC',fontweight='bold')
plt.grid(True)
plt.show()

# Find the lag with the lowest AIC value
optimal_lag = lags[np.argmin(aic_values)]
print("Optimal Lag (Time Delay):", optimal_lag)
```

```python
# Generate or load your time series data
# For demonstration purposes, let's generate a simple sinusoidal signal

signal = emg_bcps

# Function to embed data in higher dimensions
def embed_data(data, dim):
    embedded_data = np.array([data[i:i+dim] for i in range(len(data)-dim+1)])
    return embedded_data

# Function to calculate correlation sum
def calculate_correlation_sum(data, dimension, radius):
    n = len(data)
    correlation_sum = 0
    for i in range(n):
        for j in range(i+1, n):
            distance = np.linalg.norm(data[i] - data[j])
            if distance <= radius:
                correlation_sum += 1
    return correlation_sum / (n * (n - 1) / 2)

# Function to calculate correlation dimension
def calculate_correlation_dimension(data, max_dim, radius):
    correlation_dimensions = []
    for dim in range(1, max_dim+1):
        embedded_data = embed_data(data, dim)
        correlation_sum = calculate_correlation_sum(embedded_data, dim, radius)
        correlation_dimension = np.log(correlation_sum) / np.log(radius)
        correlation_dimensions.append(correlation_dimension)
    return correlation_dimensions

# Set parameters
max_dim = 10  # Maximum embedding dimension to consider
radius = 0.1  # Radius for calculating correlation sum

# Calculate correlation dimension for each embedding dimension
correlation_dimensions = calculate_correlation_dimension(signal, max_dim, radius)

# Plot correlation dimensions
plt.plot(range(1, max_dim+1), correlation_dimensions, marker='o')
plt.title("Correlation Dimension vs. Embedding Dimension (Biceps)",fontweight='bold')
plt.xlabel("Embedding Dimension",fontweight='bold')
plt.ylabel("Correlation Dimension",fontweight='bold')
plt.grid(True)
plt.show()

# Find the embedding dimension where the correlation dimension plateaus
optimal_dim = np.argmax(np.gradient(correlation_dimensions)) + 1
print("Optimal Embedding Dimension:", optimal_dim)
```

```python

def embed_time_series(ts, m, tau):
    """
    Embeds a time series into a higher-dimensional phase space using delay embedding.

    Parameters:
        ts (ndarray): The time series data.
        m (int): Embedding dimension.
        tau (int): Time delay.

    Returns:
        ndarray: Embedded time series.
    """
    N = len(ts)
    embedded_ts = np.zeros((N - (m - 1) * tau, m))
    for i in range(m):
        embedded_ts[:, i] = ts[i * tau:i * tau + len(embedded_ts)]
    return embedded_ts
def recurrence_plot(embedded_ts, epsilon):
    """
    Constructs a recurrence plot from an embedded time series.

    Parameters:
        embedded_ts (ndarray): Embedded time series.
        epsilon (float): Recurrence threshold.

    Returns:
        ndarray: Recurrence plot.
    """
    N = len(embedded_ts)
    recurrence_plot = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            recurrence_plot[i, j] = np.linalg.norm(embedded_ts[i] - embedded_ts[j])
    recurrence_plot = np.where(recurrence_plot < epsilon, 1, 0)
    return recurrence_plot

def calculate_rqa_measures(recurrence_plot):
    """
    Calculates recurrence quantification analysis (RQA) measures from a recurrence plot.

    Parameters:
        recurrence_plot (ndarray): Recurrence plot.

    Returns:
        tuple: Recurrence rate, determinism, laminarity, entropy, mean line length.
    """
    N = len(recurrence_plot)
    RR = np.sum(recurrence_plot) / (N * N)
    DET = np.sum(np.sum(np.tril(recurrence_plot, k=-1))) / np.sum(recurrence_plot)
    LAM = np.sum(np.sum(np.triu(recurrence_plot, k=1))) / np.sum(recurrence_plot)
    entropy = -np.sum((np.sum(recurrence_plot, axis=1) / np.sum(recurrence_plot)) * np.log(np.sum(recurrence_plot, axis=1) / np.sum(recurrence_plot)))
    L = 0
    num_lines = 0
    for k in range(1, N):
        for i in range(N - k):
            for j in range(N - k):
                if recurrence_plot[i, j] == 1:
                    if recurrence_plot[i + k, j + k] == 1:
                        L += 1
                    else:
                        num_lines += 1
                        L += 1
                        break
    if num_lines != 0:
        L /= num_lines
    return RR, DET, LAM, entropy, L


# Parameters
m = 3  # Embedding dimension
tau = 3  # Time delay
epsilon = 0.5  # Recurrence threshold

# Embed time series
embedded_ts = embed_time_series(emg_bcps, m, tau)

# Construct recurrence plot
recurrence_plot = recurrence_plot(embedded_ts, epsilon)

# Calculate RQA measures
RR, DET, LAM, entropy, L = calculate_rqa_measures(recurrence_plot)

# Print RQA measures
print("Recurrence Rate:", RR)
print("Determinism:", DET)
print("Laminarity:", LAM)
print("Entropy:", entropy)
print("Mean Line Length:", L)

# Plot recurrence plot
# Plot recurrence plot with bold recurrence points
plt.figure(figsize=(8, 8))
plt.imshow(recurrence_plot, cmap='binary', origin='lower')

# Highlight recurrence points
recurrence_indices = np.argwhere(recurrence_plot == 1)
plt.scatter(recurrence_indices[:, 1], recurrence_indices[:, 0], s=20, c='red')  # Adjust s for point size

plt.title("Recurrence Plot for reconstructed emg_biceps",fontweight='bold' )
plt.xlabel("Time",fontweight='bold')
plt.ylabel("Time",fontweight='bold')
plt.show()
```

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

# Function to generate surrogate data using phase randomization
def surrogate_data_phase_randomization(data):
    # Fourier transform
    ft = np.fft.fft(data)
    # Randomize phases
    random_phases = np.exp(1j * np.angle(np.random.permutation(ft)))
    # Inverse Fourier transform
    surrogate_data = np.real(np.fft.ifft(ft * random_phases))
    return surrogate_data

# Function to compute correlation dimension (a test statistic)
def correlation_dimension(data, embedding_dimension):
    # Embed the time series
    embedded_data = np.array([data[i:i+embedding_dimension] for i in range(len(data)-embedding_dimension+1)])
    # Compute pairwise distances
    distances = squareform(pdist(embedded_data))
    # Count pairs of points within a threshold
    num_pairs = np.sum(distances < 1)
    # Calculate correlation dimension
    correlation_dim = np.log(num_pairs) / np.log(embedded_data.shape[0])
    return correlation_dim

def calculate_surrogate_test_statistic(original_data, num_surrogates):
    """Calculate test statistic for surrogate data."""
    original_std = np.std(original_data)
    test_statistic = np.mean((original_data - np.mean(original_data)) ** 4) / (original_std ** 4)
    surrogate_test_statistics = []
    for _ in range(num_surrogates):
        surrogate_data = surrogate_data_phase_randomization(original_data)
        surrogate_std = np.std(surrogate_data)
        surrogate_test_statistic = np.mean((surrogate_data - np.mean(surrogate_data)) ** 4) / (surrogate_std ** 4)
        surrogate_test_statistics.append(surrogate_test_statistic)
    return test_statistic, surrogate_test_statistics
def surrogate_test(original_data, num_surrogates, confidence_level=0.99):
    """Perform surrogate test and determine p-value."""
    test_statistic, surrogate_test_statistics = calculate_surrogate_test_statistic(original_data, num_surrogates)
    sorted_surrogate_stats = np.sort(surrogate_test_statistics)
    idx = int((1 - confidence_level) * num_surrogates)
    critical_value = sorted_surrogate_stats[idx]
    p_value = np.mean(surrogate_test_statistics >= test_statistic)
    return p_value, critical_value


# Original time series data (replace this with your own data)
original_data = emg_bcps

# Number of surrogate data sets to generate
num_surrogates = 1000

# Embedding dimension (parameter for correlation dimension calculation)
embedding_dimension = 3

# Compute correlation dimension for original data
correlation_dim_original = correlation_dimension(original_data, embedding_dimension)

# Generate surrogate data and compute correlation dimensions
correlation_dims_surrogate = []
for _ in range(num_surrogates):
    surrogate_data = surrogate_data_phase_randomization(original_data)
    correlation_dim_surrogate = correlation_dimension(surrogate_data, embedding_dimension)
    correlation_dims_surrogate.append(correlation_dim_surrogate)

# Calculate the 99th percentile of the surrogate data
surrogate_99th_percentile = np.percentile(correlation_dims_surrogate, 99)

# Test for nonlinearity
if correlation_dim_original > surrogate_99th_percentile:
    print("Nonlinearity detected with 99% confidence.")
else:
    print("No significant evidence of nonlinearity detected.")

# Perform surrogate test
p_value, critical_value = surrogate_test(original_data, num_surrogates=1000)
print("P-value:", p_value)
print("Critical value at 99% confidence level:", critical_value)
```
